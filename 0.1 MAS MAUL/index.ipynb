```python
def stack():
  s=[]
  return s
def push(s,data):
  s.append(data)
def pop(s):
  data=s.pop()
  return(data)
def peek(s):
  return(s[len(s)-1])
def isEmpty(s):
  return (s==[])
def size(s):
  return(len(s))
```

```python
def subStack(bil1, bil2):
    s1 = stack()
    s2 = stack()
    shasil = stack()

    for ch in str(bil1):
        push(s1, int(ch))
    for ch in str(bil2):
        push(s2, int(ch))

    print("stack bil 1 =", s1)
    print("stack bil 2 =", s2)

    while not isEmpty(s1):
        a = pop(s1)
        b = pop(s2) if not isEmpty (s2) else 0

        if a < b:
            if not isEmpty(s1):
                sebelum = pop(s1)
                sebelum -= 1
                push(s1, sebelum)
                a += 10
        hasil = a - b
        print(f"pengurangan --> {a} - {b} = {hasil}")
        push(shasil, hasil)
        print("stack hasil:", shasil)
        print("stack bil 1 =", s1)
        print("stack bil 2 =", s2)
        print()

    hasil_final = 0
    faktor = 1
    while not isEmpty(shasil):
        hasil_final += pop(shasil) * faktor
        faktor *= 10

    return hasil_final

bil1 = 100
bil2 = 9
hasil = subStack(bil1, bil2)
print("hasil =", hasil)
```

```python
def schedulingResource(dataTask, targetTask):
    queue = []
    waktu = 0

    # Step 1: Masukkan semua task ke antrian
    for task in dataTask:
        queue.append([task[0], task[1]])  # ['Task 1', 4]

    # Step 2: Proses per detik
    while True:
        print("queue :", [t[0] for t in queue])
        
        current = queue.pop(0)
        current[1] -= 1  # dikerjakan 1 detik
        waktu += 1

        # Kalau belum selesai, masukin lagi ke belakang
        if current[1] > 0:
            queue.append(current)
        else:
            # Kalau task yang selesai adalah target → selesai
            if current[0] == targetTask:
                break
        # Kalau task lain yang selesai → lanjut

    return waktu
dataTask = [['Task 1',4], ['Task 2',7], ['Task 3',1]]
ind = 'Task 2'
waktu = schedulingResource(dataTask, ind)
print('waktu yang dibutuhkan untuk', ind, ':', waktu, 'detik')
```
