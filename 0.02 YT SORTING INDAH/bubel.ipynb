```python
data = [6, 10, 12, 100, 1, 0, 23, 45, 7, 8, 9]
n = len(data)

i = 0
swapped = True

while i < n - 1 and swapped:
    print(f"\nIterasi ke --> {i+1}")
    print(f"Jumlah iterasi --> {n-1-i}")
    
    swapped = False

    for j in range(n - 1 - i):
        print(f"{j+1} = {data}")
        if data[j] > data[j+1]:
            data[j], data[j+1] = data[j+1], data[j]
            swapped = True
    
    i += 1
```
