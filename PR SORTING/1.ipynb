```python
data=[4,3,2,1]
for j in range(1, len(data)):
    print(f"LITERASI KE - {j} \nDATA = {data} KE - {j}")
    key = data[j]
    i = j-1
    while i >=0 and data[i] > key:
        data[i+1] = data[i]
        i -= 1
        data[i+1] = key
    print(f"DATA SEKARANNG ADALAH = {data}\n")
```

ASCENDING DARI KIRI KE KANAN

```python
data = [10, 8, 90, 87, 89, 65, 60, 99, 100, 1, 3]
for j in range(1, len(data)):
    swapped = False
    iterasi_baru = []
    
    for i in range(len(data) - j):  
        if data[i] > data[i + 1]:
            data[i], data[i + 1] = data[i + 1], data[i] 
            swapped = True
        iterasi_baru.append(str(data))

    if not swapped:
        print(f"Data sudah terurut setelah iterasi ke-{j}")
        break

    print(f"literasi ke-{j}")
    for baris in iterasi_baru:
        print(baris)
    print("")
```

ASCENDING DARI KANAN KE KIRI

```python
data = [10, 8, 90, 87, 89, 65, 60, 99, 100, 1, 3]
for j in range(1, len(data)):
    swapped = False
    iterasi_baru = []
    
    for i in range(len(data) - 1, j - 1, -1):
        if data[i] < data[i - 1]:  
            data[i], data[i - 1] = data[i - 1], data[i]  
            swapped = True
        iterasi_baru.append(str(data))

    if not swapped:
        print(f"Data sudah terurut setelah iterasi ke-{j}")
        break

    print(f"literasi ke-{j}")
    for baris in iterasi_baru:
        print(baris)
    print("")
```

```python
data = [10, 8, 90, 87, 89, 65, 60, 99, 100, 1, 3]
print("Bubble Sort Descending")

for j in range(1, len(data)):
    swapped = False
    iterasi_baru = []

    for i in range(len(data) - j):
        if data[i] > data[i + 1]:
            data[i], data[i + 1] = data[i + 1], data[i]
            swapped = True
            iterasi_baru.append(str(data))

    if swapped:
        print(f"literasi ke-{j}")
        for baris in iterasi_baru:
            print(baris)
        print("")
```

```python
data = [10, 8, 90, 87, 89, 65, 60, 99, 100, 1, 3, 99, 900, 809, 5, 7]
print("Selection Sort Ascending")
for i in range(len(data)):
    min_idx = i
    for j in range(i + 1, len(data)):
        if data[j] < data[min_idx]:
            min_idx = j
    if min_idx != i:
        data[i], data[min_idx] = data[min_idx], data[i]
        print(f"iterasi ke-{i+1}: {data}")
    else:
        print(f"iterasi ke-{i+1}: (tidak ada pertukaran)")
```

```python
data = [10, 8, 90, 87, 89, 65, 60, 99, 100, 1, 3]
print("Selection Sort Descending")
for i in range(len(data)):
    max_idx = i
    for j in range(i + 1, len(data)):
        if data[j] > data[max_idx]:
            max_idx = j
    if max_idx != i:
        data[i], data[max_idx] = data[max_idx], data[i]
        print(f"iterasi ke-{i+1}: {data}")
    else:
        print(f"iterasi ke-{i+1}: (tidak ada perubahan, dihentikan)")
```
